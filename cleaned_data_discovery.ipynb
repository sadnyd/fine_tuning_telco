```python
from datasets import load_from_disk
from datasets import DatasetDict

# Load your freshly scrubbed local dataset
train_split = load_from_disk("./cleaned_telco_dataset")

# Reconstruct a DatasetDict format so the rest of your script runs unmodified

ds = DatasetDict({"train": train_split})
print(f"Cleaned dataset loaded: {ds}")
```

```python
import pandas as pd
from collections import Counter
import re

word_counts = Counter()

for text in ds["train"]["instruction"]:
    words = re.findall(r"\b\w+\b", text.lower())
    word_counts.update(words)

df_words = pd.DataFrame(
    word_counts.items(),
    columns=["word", "count"]
).sort_values("count", ascending=False)

df_words.head(50)
```
