```python
import pandas as pd
from collections import Counter
import re

word_counts = Counter()

for text in ds["train"]["instruction"]:
    words = re.findall(r"\b\w+\b", text.lower())
    word_counts.update(words)

df_words = pd.DataFrame(
    word_counts.items(),
    columns=["word", "count"]
).sort_values("count", ascending=False)

df_words.head(50)
```

```python
from collections import Counter

intent_counts = Counter(ds["train"]["intent"])

print("Total unique intents:", len(intent_counts))
print()

for intent, count in intent_counts.most_common():
    print(f"{intent:<40} {count}")
```

```python
from collections import Counter
import re

for intent in sorted(set(ds["train"]["intent"])):

    words = Counter()

    rows = ds["train"].filter(
        lambda x: x["intent"] == intent
    )

    for text in rows["instruction"]:
        tokens = re.findall(r"\b\w+\b", text.lower())
        words.update(tokens)

    print(f"\n{'='*80}")
    print(intent)
    print(f"{'='*80}")

    for word, count in words.most_common(20):
        print(f"{word:<20} {count}")
```

```python
categories = ds["train"]["category"]

unique_categories = sorted(set(categories))

print("Number of unique categories:", len(unique_categories))
print()

for category in unique_categories:
    print(category)
```

```python
from collections import Counter
import re

for category in sorted(set(ds["train"]["category"])):

    words = Counter()

    rows = ds["train"].filter(
        lambda x: x["category"] == category
    )

    for text in rows["instruction"]:
        tokens = re.findall(r"\b\w+\b", text.lower())
        words.update(tokens)

    print(f"\n{'='*80}")
    print(category)
    print(f"{'='*80}")

    for word, count in words.most_common(30):
        print(f"{word:<30} {count}")
```

```python
tags = ds["train"]["tags"]

unique_tags = sorted(set(tags))

print("Number of unique categories:", len(unique_tags))
print()

for tag in unique_tags:
    print(tag)
```

```python

word_counts = Counter()

for text in ds["train"]["response"]:
    words = re.findall(r"\b\w+\b", text.lower())
    word_counts.update(words)

df_words = pd.DataFrame(
    word_counts.items(),
    columns=["word", "count"]
).sort_values("count", ascending=False)

df_words.head(50)
```
